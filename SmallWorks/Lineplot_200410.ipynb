```
from matplotlib import pyplot as plt
import numpy as np
```

```
#구성원 점수 그래프 그리기

student_score = [
[1, 90, 80, 85, 0, 0.0, 0, 0],
[2, 88, 85, 90, 0, 0.0, 0, 0],
[3, 32, 85, 90, 0, 0.0, 0, 0],
[4, 42, 75, 93, 0, 0.0, 0, 0],
[5, 82, 86, 88, 0, 0.0, 0, 0]
]

X = ['kor','math','eng']
for i in range(len(student_score)):
  Yi = student_score[i][1:4]
  plt.plot(X,Yi,'o--')
plt.xlabel("subject")
plt.ylabel("grade")
plt.title("GradeGraph")
plt.legend(['st1', 'st2', 'st3', 'st4', 'st5'])
plt.show()
```

##기본 이론

```
x = np.arange(1,11)     #ndarray
y = 2 * x + 5     #ndarray에 산술연산이 적용되면 스칼라배로 연산됨
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y,"bo")      #plot은 함수여서 show 함수로 그래프를 호출해야함
plt.show()
```

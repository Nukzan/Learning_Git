#Plus Work

##kbs main news title and links

```
kbsmain = requests.get('http://news.kbs.co.kr/common/main.html')
kbsmsoup = BeautifulSoup(kbsmain.content,'html.parser')
kbstit = kbsmsoup.find_all('em',class_='tit')


# Main title list

text_list = []
for text in kbstit:
  text_list.append(text)

text_list

for title in text_list:
  print(title.text)
```

##naver main news title and links

```
nvmain = requests.get('https://news.naver.com/')
nvmsoup = BeautifulSoup(nvmain.content,'html.parser')
nvtit = nvmsoup.find_all('ul',class_='hdline_article_list')

# Main title list

text_list1 = []
for text1 in nvtit:
  text_list1.append(text1)

for title1 in text_list1:
  print(title1.text)
  print(title1.select('a[href]'))
```

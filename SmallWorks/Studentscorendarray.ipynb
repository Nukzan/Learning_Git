## 기본 과제

```
import numpy as np
```

```
#{'num':'1', 'kor':90,'eng':80,'math':85,'total':0,'avg':0.0,'min':80,'max':90}
students_score = [
                  [1,90,80,85,0,0.0,0,0],
                  [2,88,85,90,0,0.0,0,0],
                  [3,32,85,90,0,0.0,0,0],
                  [4,42,75,93,0,0.0,0,0],
                  [5,82,86,88,0,0.0,0,0]
]
```

```
# 각 구성원 total, avg 필드 각 생성 및 값 구하기

#total
ndarray = np.array(students_score)
ndarray[:,4] = np.sum(ndarray[:,1:4],axis=1)

#avg
ndarray[:,5] = np.mean(ndarray[:,1:4],axis=1)

#min
ndarray[:,6] = np.min(ndarray[:,1:4],axis=1)

#max
ndarray[:,7] = np.max(ndarray[:,1:4],axis=1)

ndarray
```

```
#학급 평균 구하기
class_avg = np.mean(ndarray[:,5])
class_avg
```

```
#과목별 평균 구하기
print("kor_avg =", np.mean(ndarray[:,1]))
print("eng_avg =", np.mean(ndarray[:,2]))
print("math_avg =", np.mean(ndarray[:,3]))

```

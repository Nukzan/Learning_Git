**문제 풀이 순서**\
1) 2명의 정보 update\
2) 각 구성원 total, avg 함수\
3) 학급 평균 구하는 함수\
4) 과목별 평균 구하는 함수

```
#변수 선언
students_score = [
                  {'num':1, 'name':'김이나','kor':90,'eng':80,'math':85},
                  {'num':2, 'name':'홍상지','kor':90,'eng':85,'math':90},
                  {'num':3, 'name':'최지원','kor':80,'eng':80,'math':80}
]
```

```
#학생 목록 (딕셔너리) 추가
students_score.append({'num':4, 'name':'유수현','kor':75,'eng':95,'math':100})
students_score.append({'num':5, 'name':'강희주','kor':95,'eng':100,'math':85})
students_score
```

```
#각 학생의 total, avg 채워 넣는 함수
def each_Total_Avg(list):
  for i in range(len(list)):
    list[i]['total'] = list[i]['kor']+list[i]['eng']+list[i]['math']
    list[i]['avg'] = list[i]['total'] / 3
  print(list)

each_Total_Avg(students_score)
```

```
#학급 total 구하는 함수
classTotal = []

def class_total(list):
    for i in range(len(list)):
      classTotal.append(list[i]['total'])
    total = sum(classTotal) / len(classTotal)
    print("해당 학급의 총 점수는 ", total,"점 입니다.")

class_total(students_score)

```

```
#학급 avg 구하는 함수
classAvg = []

def class_Avg(list):
    for i in range(len(list)):
      classAvg.append(list[i]['avg'])
    AVG = round((sum(classAvg) / len(classAvg)),1)
    print("해당 학급의 전체 평균 점수는 ", AVG,"점 입니다.")

class_Avg(students_score)
```

```
#과목별 평균 구하는 함수
kor_avgList = []
eng_avgList = []
math_avgList = [] 

def sbj_Avg(list):
    for i in range(len(list)):
      kor_avgList.append(list[i]['kor'])
      eng_avgList.append(list[i]['eng'])
      math_avgList.append(list[i]['math'])
    kor_AVG = round((sum(kor_avgList) / len(kor_avgList)),1)
    eng_AVG = round((sum(eng_avgList) / len(eng_avgList)),1)
    math_AVG = round((sum(math_avgList) / len(math_avgList)),1)
    print("해당 학급의 국어 평균 점수는 ", kor_AVG,"점 입니다.")
    print("해당 학급의 영어 평균 점수는 ", eng_AVG,"점 입니다.")
    print("해당 학급의 수학 평균 점수는 ", math_AVG,"점 입니다.")


sbj_Avg(students_score)
```

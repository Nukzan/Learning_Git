```
student_list = [
 [1, 90, 80, 85, 0, 0.0, 0],
 [2, 90, 85, 90, 0, 0.0, 0],
 [3, 80, 80, 80, 0, 0.0, 0]
]
```

```
student_dictionary = [
{'num':'1', 'name':'김철수','kor':90, 'eng':80, 'math':85, 'total':0, 'avg':0.0, 'order':0 },
{'num':'2', 'name':'박제동','kor':90, 'eng':85, 'math':90, 'total':0, 'avg':0.0, 'order':0 },
{'num':'3', 'name':'홍길동','kor':80,' eng':80, 'math':80, 'total':0, 'avg':0.0, 'order':0 }
]
```

```
student_list, student_dictionary
```

#Update

```
student_list[1], student_dictionary[2]
```

```
# Row update
student_list[1] = [1,95,89,88,0,0.0,0]
student_dictionary[2] = {' eng': 83,  'avg': 0.0,  'kor': 92,  'math': 93,  'name': '송이나',  'num': '1',  'order': 0,
  'total': 0}
print(student_list[1])
print(student_dictionary[2])
```

```
# Column update
student_list[1][2] = 96
student_dictionary[2]['eng']= 94
student_list[1], student_dictionary[2]
```

# Insert & Append

*** Insert ***

```
student_list. insert(0,[1,90,80,85,0,0.0,0])
student_dictionary.insert(len(student_dictionary), {'avg': 0.0,'eng': 58,'kor': 88,'math': 100,'name': '정유미','num': '4',
   'order': 0,'total': 0})
student_list, student_dictionary
```

```
student_list[1].insert(0,2)
student_dictionary[1]['name']='김미리'
student_list, student_dictionary
```

```
#딕셔너리 insert에서 키값을 파라미터로 넣는 게 가능한가
student_dictionary[1].insert('temp','바람개비')
student_dictionary
```

# For문

```
# in 다음에 오는 대상의 가장 외부가 리스트로 싸여있어야 동작함
# 나오는 순서는 List의 행의 순서로 나옴
for item in student_list:
  print(item, type(item))
```

```
# 딕셔너리도 가능함 (순서로 리스팅이 가능하므로...?)
for item2 in student_dictionary:
  print(item2, type(item2))
```

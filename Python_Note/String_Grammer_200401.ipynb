```
A = "korean culture"
A.count("ul") #string 내장함수 count /  함수() 안에 내용을 넣지 않으면 ERROR가 뜸
```

```
A.find("e") #해당 내용이 나온 위치(색인) 중 첫 번째 위치만 반환 -> 맹점!
```

```
A.find("K")
#대상 범위 내 해당 내용이 없을 때는 -1 반환 -> 내가 원하는 글자가 있는지 없는지를 찾을 때, -1 이상인지 확인하면 가능
```

*** Temporary Array Creation ***

```
# empty
x1 = np.empty([3,2], dtype = int)

#zeros
x2 = np.zeros((2,2), dtype = [('x','i4'),('y','i4')])

#ones
x3 = np.ones([2,2], dtype = int)

#asarray
x4 = (1,2,3)
a = np.asarray(x4)

#arragne
x5 = np.arange(10,20,2)

print(x1,type(x1))
print(x2, type(x2))
print(x3, type(x3))
print(a,type(a))
print(x5,type(x5))

plt.plot(x4)
plt.show()
```

*** List and Ndarray ***

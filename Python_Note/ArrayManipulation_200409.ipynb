#Manipulation

## concatenate

```
import numpy as np
```

```
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
print(np.concatenate((a,b)))
print(np.concatenate((a,b),axis=0))
print(np.concatenate((a,b),axis=1))
```

```
a1 = np.array([[1,2],[3,4]])
b1 = np.array([[5,6],[7,8],[9,10]])
print(np.concatenate((a1,b1)))
print(np.concatenate((a1,b1),axis=0))
print(np.concatenate((a1,b1),axis=1))

#Error 이유 : 매트릭스는 박스가 만들어져야 하는데, 형태(차원)이 맞지 않으면 만들 수 없음
```

##split

```
a2 = np.arange(9)
b2 = np.split(a2,3)

a2, type(a2), b2, type(b2)
```

##resize

```
a3 = np.array([[1,2,3],[4,5,6]])
resz = np.resize(a3,(3,2))
resz, resz.shape
```

##insert

```
a4 = np.array([[1,2],[3,4],[5,6]])
np.insert(a4, 1,[11,12],axis=0), np.insert(a4,2,11,axis=1)
print(a4)
```

##delete

```
a5 = np.arange(12).reshape(3,4)
print(a5)
print(np.delete(a5,5))      #인덱스의 값을 삭제
print(np.delete(a5,1,axis=1))     #인덱스 열을 삭제 (axis=0 : 인덱스 행을 삭제)
```

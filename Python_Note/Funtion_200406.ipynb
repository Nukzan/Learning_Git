*** Parameter ***

```
def print_return_tuple(arg1, *vartuple):
  print("inside arg1 is : ", arg1)
  for var in vartuple:
    print("Inside vartuple is :", var)
  return vartuple;

print_return_tuple(10)      #for문에서 vartuple 자체가 없어서 안돌아가므로, 함수의 첫 문장만 실행되고 에러는 안남

#함수 파라미터와 계산식이 매칭되므로, 70은 vartuple에 들어가지 않음 -> 해당 줄의 결과값에서 70이 제외됨
#파라미터에 들어왔다고 무조건 리턴으로 되는 것은 아님
out_tuple = print_return_tuple(70,60,50)
print(out_tuple)      
```

```
#일반적인 print와 같은 방식의 함수
def print_return_tuple2(*vartuple2):
  for var2 in vartuple2:
    print("Inside vartuple is :", var2)
  return vartuple2;

out_tuple2 = print_return_tuple2(70,60,50)
print(out_tuple2)
```

*** Lambda ***\
반복적으로 사용하지만,\
함수 만들기에는 비효율적일 때\
단순화된 함수 정도로 사용
- 입력 parameter만 있으면 됨
- 파라미터가 적용되는 영역은 ':'으로 정함
- (+) kotlin 에서는 다르게 적용되는 lambda

```
sumLam = lambda arg1, arg2 : arg1 + arg2;
print("Value of total : ", sumLam(10,20))
```

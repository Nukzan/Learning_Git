Data Preprocessing

- how to fill missing values (NA)

```
#fill NA

df = pd.DataFrame(np.random.randn(3,3), index=['a','c','e'],columns = ['one', 'two', 'three'])
df
```

```
df = df.reindex(['a','b','c'])
df
```

```
df = df.fillna(df.mean())
df
```

```
#drop NA

df1 = pd.DataFrame(np.random.randn(5,3),index=['a','c','e','f','h'],columns = ['one','two','three'])
df1
```

```
df1 = df.reindex(['a','b','c','d','e','f','g','h'])
df1
```

```
df1 = df1.dropna()
```

```
#replace

df2 = pd.DataFrame({'one':[10,20,30,40,50,2000],'two':[1000,0,30,40,50,60]})
df2
```

```
df2 = df2.replace({1000:10,2000:60})
df2
```

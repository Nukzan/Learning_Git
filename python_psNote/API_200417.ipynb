##crawling from naver finance for all pages - looping

```
import time
```

```
url1 = 'https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001&date=20200417&page={}'
for x in range(0,500):
  navns = url1.format('000000',x)
  res_nav = requests.get(navns)
  soup_nav = BeautifulSoup(res_nav.content)
  title = soup_nav.find('div',id ='main_content')
  print(title)
```

*** while 문 ***

```
#while문 넣기, 조건을 만족하면 멈추기
while 1:
  if score >= 90:
   letter = "A"
   break
  elif score >= 80:
   letter = "B"
  elif score >= 70:
   letter = "C"
  elif score >= 60:
   letter = "D"
  else:
   letter = "F"
print(letter)
```

```
#while에 조건을 넣어 반복문 끝내기
score = int(input("What's your score?"))
while (score > 50):
  score = int(input("What's your score?"))
  if score >= 90:
   letter = "A"
  elif score >= 80:
   letter = "B"
  elif score >= 70:
   letter = "C"
  elif score >= 60:
   letter = "D"
  else:
   letter = "F"
print(letter)
```

```
#while에 조건을 넣어 반복문 끝내고 while 조건이 맞지 않는 경우 else 사용
#입력받는 변수가 없는 경우 count 변수를 사용함
count = 0
while (count < 5):
  score = int(input("What's your score?"))
  if score >= 90:
    letter = "A"
  elif score >= 80:
    letter = "B"
  elif score >= 70:
    letter = "C"
  elif score >= 60:
    letter = "D"
  else :
    letter = "Under grade. You need to study more."
  print(letter)
  count = count + 1
else : 
  print("Today's calculation work is ended")
```

```
#len을 활용해서 변수의 길이만큼만 값을 도출
movies = ['the holy grail', 'the life of brian', 'the meaning of life']
for each_movie in movies:
  print(each_movie)
```

```
#이미 입력받는 변수가 있는 경우에는 그 변수를 활용해서 while 조건문을 작성하는 게 좋음(깔끔함)
score = int(input("What's your score?"))
while (score > 20):
  if score > 100 and score < 0:
    letter = "Wrong Score Range. Try again."
  if score >= 90:
    letter = "A"
  elif score >= 80:
    letter = "B"
  elif score >= 70:
    letter = "C"
  elif score >= 60:
    letter = "D"
  else :
    letter = "F"
  print(letter)
  score = int(input("What's your score?"))
else : 
  print("You didn't pass the subject. Need to study hard.")
```

*** TRY - For 구현과 이해 ***

첫 번째 for문은 행을 만들고,\
두 번째 for문은 각각 쉘을 채움(열방향)

```
#기초 테스트
for num in range(5,10):
  for num2 in range(2,num):
    print(num, num2)
```

```
for num3 in range(10,20):
  for num4 in range(2,num3):
    if num3 % num4 == 0:
      num5 = num3 / num4
      print('%d equals %d * %d' % (num3,num4,num5))
      break
#  else :
#    print(num3, 'is a prime number')
```

```
#zip()
seq1 = ['foo','bar','baz']
seq2 = ['one','two','three']
seq3 = ['apple','orange','banana']
for a,b,c in zip(seq1, seq2, seq3):
  print('a:{},b:{},c:{}'.format(a,b,c,))
```

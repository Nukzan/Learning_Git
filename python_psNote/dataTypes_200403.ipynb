*** (사용자함수) 비교함수 cmp ***

```
#return의 a와 b 비교는 T/F의 결과로 나옴
#T와 F는 각각 1과 0으로 연산이 가능함
def cmp(a,b):
  return (a > b) - (a < b)
```

```
tuple1, tuple2 = (123,'xyz'), (456,'abc')
```

```
cmp(tuple1,tuple2)
```

```
print(tuple1 > tuple2)
```

```
print(tuple1 < tuple2)
```

```
print(False - True)
```

```
cmp(tuple2,tuple1)
```

```
#tuple + tuple 의 경우 뒤의 내용이 뒤로 추가됨
tuple3 = tuple2 + (786,);
cmp(tuple2, tuple3)
```

```
tuple3
```

```
#앞의 값이 동일하면 순차적으로 뒤를 비교
#(456, 'abc') < (456, 'abc', 786)
print(tuple2 < tuple3)
```

```
#(456, 'abc') > (456, 'abc', 786)
print(tuple2 > tuple3)
```

```
#문자는 a->x 순으로 값이 커짐
('abc') < ('xyz')
```

*** Dictionary 함수 ***

```
dic1 = {'name':'zara', 'age':'7','class':'first'}
```

```
print("dic1['name'] : ", dic1['name'])
```

```
len(dic1), str(dic1), type(dic1)
```

```
dic2 = dic1.copy()
```

```
#dic1과 dic2는 갖고 있는 내용은 갖지만, 사본과 원본이므로 둘이 완전히 "같은" 것이 아님
dic1 is dic2
```

```
#get 함수의 파라미터 "get(찾을 키, 해당 키가 없다면 보고자 하는 값)"
dic1.get('age')
```

```
#keys 함수는 리스트로 결과값 반환
#list 형태이기 때문에 for문으로 돌릴 수 있음 (반복문)
dic1.keys()
```

```
#새로운  키 값을 추가하는 함수 setdefault
#새 키를 추가하면 value값은 None으로 들어감
dic1.setdefault('Age')
```

```
dic1
```

```
#setdefault(새 키, 들어갈 value)
dic1.setdefault('grade',"A")
dic1
```

```
#기존에 존재하는 키를 넣을 경우, value는 변경되지 않음
dic1.setdefault('grade',"B")
dic1
```

```
dic1.get('Grade', "nothing")
```

```
#기존 + 연산자로 하던 작업을 update을 통해 처리
dic3 = {'sex':'female'}
dic1.update(dic3)
```

```
dic1
```

```
dic1.values()
```

```
dic1.items()
```

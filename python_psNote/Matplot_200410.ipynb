```
from matplotlib import pyplot as plt
import numpy as np
```

```
#구성원 점수 그래프 그리기

student_score = [
[1, 90, 80, 85, 0, 0.0, 0, 0],
[2, 88, 85, 90, 0, 0.0, 0, 0],
[3, 32, 85, 90, 0, 0.0, 0, 0],
[4, 42, 75, 93, 0, 0.0, 0, 0],
[5, 82, 86, 88, 0, 0.0, 0, 0]
]

X = ['kor','math','eng']
for i in range(len(student_score)):
  Yi = student_score[i][1:4]
  plt.plot(X,Yi,'o--')
plt.xlabel("subject")
plt.ylabel("grade")
plt.title("GradeGraph")
plt.legend(['st1', 'st2', 'st3', 'st4', 'st5'])
plt.show()
```

###Histogram

```
n = 5 + np.random.randn(1000)
m = [m for m in range(len(n))]
plt.bar(m,n)
plt.title("raw data")
plt.show()
```

```
plt.hist(n, bins=20)
plt.title("histogram")
plt.show()
```

```
plt.hist(n, cumulative=True, bins =20)
plt.title("cumulative histogram")
plt.show()
```

###iris 데이터

```
from sklearn import datasets

iris = datasets.load_iris()
#iris['data']
print(iris['feature_names'])
print(iris['target_names'])
```

```
#iris class(종류)별 그래프 그리기
#종류 3종 : 'setosa', 'versicolor', 'virginica'
#속성 : 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'

#속성으로 구분하기
sepal_length = iris['data'][:,0]
sepal_width = iris['data'][:,1]
petal_length = iris['data'][:,2]
petal_width = iris['data'][:,3]

iris_class = iris['feature_names']
iris_target = iris['target']


#속성을 클래스 별로 구분하기
for i in range(len(iris_class)):
  s_Li = sepal_length[iris_target ==i]
  s_Wi = sepal_width[iris_target ==i]
  p_Li = petal_length[iris_target ==i]
  p_Wi = petal_width[iris_target ==i]
  plt.scatter(s_Li,s_Wi, marker='v')
  plt.scatter(p_Li,p_Wi,marker='o')
plt.title('Iris Scatter Plot')
plt.legend(['set_Sep','set_Pet','ver_Sep','ver_Pet','vir_Sep','vir_Pet'])
plt.show()
```

**문제 풀이 순서**\
1) 2명의 정보 update\
2) total, avg 필드 생성\
3) 각 구성원 total, avg 값 구하기\
4) 학급 평균 구하기\
5) 과목별 평균 구하기

```
#변수 선언
students_score = [
                  {'num':1, 'name':'김철수','kor':90,'eng':80,'math':85},
                  {'num':2, 'name':'박제동','kor':90,'eng':85,'math':90},
                  {'num':3, 'name':'홍길동','kor':80,'eng':80,'math':80}
]
```

```
# 1) students_score 리스트에 '류수연','김혜수' 딕셔너리 추가
students_score.append({'num':4, 'name':'김혜수','kor':100,'eng':45,'math':100})
students_score.append({'num':5, 'name':'류수연','kor':95,'eng':100,'math':94})
students_score
```

```
# 2) 각 학생 dict에 'total', 'avg' 필드 생성
for addtotal in students_score:
  addtotal.setdefault('total')
  addtotal.setdefault('avg')
students_score
```

```
# 3) 각 학생마다 'total','avg' 값 계산
count = 0
while (count <len(students_score)):
  students_score[count]['total'] = students_score[count]['kor'] + students_score[count]['eng'] + students_score[count]['math']
  students_score[count]['avg'] = round(((students_score[count]['kor'] + students_score[count]['eng'] + students_score[count]['math']) / 3 ),1)
  count = count + 1
students_score
```

```
# 4) 학급 평균 구하기 : 학생들의 avg 키의 value 값들을 더해서 학생 수(len)로 나눔
#avgList : 학생 평균점수 리스트를 만든다
avgList = []
for students in students_score:
  avgList.append(students.get('avg'))
print(avgList)

#ClassAvg : 평균점수 리스트의 합을 구한 후 요소 수로 나눔
ClassAvg = sum(avgList) / len(avgList)
ClassAvg
```

```
# 5) 과목별 평균 구하기 : 학생들의 각각 과목별 값을 구한 후, 과목별로 더해서 학생 수(len)로 나눔
#korGrd : kor 점수 리스트 / KorAvg : kor 점수 평균
korGrd = []
for kor_score in students_score:
  korGrd.append(kor_score.get('kor'))
print(korGrd)
KorAvg = sum(korGrd) / len(korGrd)
KorAvg
```

```
#engGrd : eng 점수 리스트 / engAvg : eng 점수 평균
engGrd = []
for eng_score in students_score:
  engGrd.append(eng_score.get('eng'))
print(engGrd)
engAvg = sum(engGrd) / len(engGrd)
engAvg
```

```

#mathGrd : math 점수 리스트 / mathAvg : math 점수 평균
mathGrd = []
for math_score in students_score:
  mathGrd.append(math_score.get('math'))
print(mathGrd)
mathAvg = sum(mathGrd) / len(mathGrd)
mathAvg
```

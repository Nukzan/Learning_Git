```
#기사 중 기호를 스페이스로 대체 : replace 메서드 중첩 사용
#아래에서 한 번에 실행하고 싶지만, replace가 string에서만 가능해서 최상단에서 사용
news_replace = news_str.replace('.',' ').replace('<',' ').replace('[',' ').replace('>',' ').replace(']',' ').replace('·',' ').replace('“',' ').replace('”',' ').replace('=',' ').replace('▲',' ')
```

```
news_replace
```

```
#빈칸으로 단어 나눔
news_splt = news_replace.split(' ')
news_splt
```

```
#2번 셀에서 삽입 된 빈칸 제거
news_list = ' '.join(news_splt).split()
```

```
count = 0
verb_list = []      #동사 리스트 선언
noun_list = []      #명사 리스트 선언
anonymous_list = []     #기타 리스트 선언
while (count < len(news_list)):
  #해당 값에 조건이 해당하는 지 파악하여, 동사 리스트 생성
  if (("다" in news_list[count]) or ("다는" in news_list[count]) or ("아냐" in news_list[count]) or ("?" in news_list[count]) or ("되"in news_list[count])):
    verb_list.append(news_list[count])
  elif (("적"in news_list[count]) or ("로"in news_list[count]) or (("더"in news_list[count]) and (len(news_list[count])==1))
  or ("같은"in news_list[count]) or ("하지만"in news_list[count]) or ("록"in news_list[count]) or (("수"in news_list[count]) and (len(news_list[count])==1)) or ("라고"in news_list[count]) or ("지만"in news_list[count])):
    anonymous_list.append(news_list[count])
  else:
    noun_list.append(news_list[count])
  count = count + 1
print('verb_list =', verb_list)
print('noun_list =', noun_list)
print('anonymous_list =', anonymous_list)
#----------------------------------------------------------
#현재 문제점 : 줄바꿈을 하면서 맨 끝 글자와 맨 앞 글자가 붙어서 나옴
#수동으로 원문 수정하는 것 없이 자동적으로 띄어쓰는 방법을 아직 못찾음 
#ex : noun_list의 '도출타시도'
#----------------------------------------------------------
```

(1) for문 사용하여 출력\
(2) 출력 function으로 전환

```
#변수 선언
movies = ["the holy grail", "the life of brian", "the meaning of life"]

#for문 사용하여 출력
for i in range(len(movies)):
  print(movies[i])
```

```
#function 작성
def printList(list):
  for i in range(len(list)):
    print(list[i])
  return;

printList(movies)
```

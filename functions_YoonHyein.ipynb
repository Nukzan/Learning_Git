```
def sumFunc(par1, par2):
  sumResult = int(par1) + int(par2)     #int 는 캐스팅하는 함수
  print("Inside the function : ", sumResult)
  return sumResult;

fst, scd = input("input two number?(put comma(,) between numbers)").split(',')
final = sumFunc(fst, scd)
print("Outside the function : ", final)
```

```
#추가 사칙연산 적용하기
def restFunc(par1, par2):
  minusResult = int(par1) + int(par2)
  mutiResult = int(par1) * int(par2)
  divResult = int(par1) / int(par2)
  print("Inside the function : ", minusResult ,"&", mutiResult,"&",divResult)
  return minusResult,mutiResult, divResult ;

fst, scd = input("input two number?(put comma(,) between numbers)").split(',')
final2 = restFunc(fst, scd)
print("Outside the function : ", final2)
```
